```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 8]

def fcentered(f, x, h):
  return ( f(x+h) - f(x-h) ) / (2*h)

def foobar(x):
  return x**4

fprime_exact = 32 #Because 4*x**3 = 32 @ x = 2

h = np.logspace(-16, 0, 1001)

fprime_approx = fcentered(foobar, 2, h)

print("approx: ", fprime_approx)
print("exact: ", fprime_exact)



ET = abs(fprime_exact - fprime_approx)

plt.clf; plt.plot(h, ET, 'r-')

plt.loglog()

plt.xticks(10.**np.arange(-16, 1), fontsize = 15)

plt.yticks(10.**np.arange(-12, 2), fontsize = 15)

plt.grid()

ET.argmin()

optimalh = h[ET.argmin()]

print(h[h == 1e-4])

np.where(h == 1e-4)

np.where(h == 1e-2)

plt.plot(h[750:876], ET[750:876], 'b-')


c = np.polyfit(np.log(h[750:876]), np.log(ET[750:876]), 1)

c

cR = np.polyfit(np.log(h[:450]), np.log(ET[:450]), 1)

cR


plt.plot(h[:450], ET[:450], 'm-', lw = 3)

plt.plot(h[:450], np.exp(-34.14957233)*h[:450]**(-0.99760735), 'k-', lw = 3)



print('\n')
print('The parameters for the linear regression of the blue line are:', c)
print('So, the slope for the blue line is 2.00000213')
print('\n')
print('The parameters for the linear regression of the black line are:', cR)
print('So, the slope for the black line is -0.99760735')
print('\n')

print('\n')
print('The optimal choice of h when approximating the derivative of the function f(x) = x^4 at x = 2 is:', optimalh)
print('\n')
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 8]

def frenchcentered(f, x, y, h):
  return ( f(x + h, y) + f(x - h, y) + f(x, y + h) + f(x, y - h) - 4*f(x, y) ) / (h**2)

def fooness(x, y):
  return np.sin(x+y)

fprime_exact = (-(2**0.5)) #Because laplace of sin(x+y) is -2sin(x+y) and @ x = pi/8 and y = pi/8 that is (-(2**0.5)) and this seems to produce a graph that looks correct

h = np.logspace(-7.5, 0, 1001)

fprime_approx = frenchcentered(fooness, (np.pi / 8), (np.pi / 8), h)

print("approx: ", fprime_approx)
print("exact: ", fprime_exact)


ET = abs(fprime_exact - fprime_approx)

plt.clf; plt.plot(h, ET, 'r-')

plt.loglog()

plt.xticks(10.**np.arange(-8, 1), fontsize = 15)

plt.yticks(10.**np.arange(-18, 3), fontsize = 15)

plt.grid()

ET.argmin()

optimalh = h[ET.argmin()]


plt.plot(h[650:876], ET[650:876], 'b-')


c = np.polyfit(np.log(h[650:876]), np.log(ET[650:876]), 1)



cR = np.polyfit(np.log(h[:450]), np.log(ET[:450]), 1)




plt.plot(h[:450], ET[:450], 'm-', lw = 3)

plt.plot(h[:450], np.exp(-36.3240438)*h[:450]**(-1.99077262), 'k-', lw = 3)



print('\n')
print('The parameters for the linear regression of the blue line are:', c)
print('So, the slope for the blue line is 1.99993043')
print('\n')
print('The parameters for the linear regression of the black line are:', cR)
print('So, the slope for the black line is -1.99077262')
print('\n')

print('\n')
print('The optimal choice of h when approximating the derivative of the function f(x, y) = sin(x+y) at x = pi/8 and y = pi/8 is:', optimalh)
print('\n')
```
